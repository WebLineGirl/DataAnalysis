```python
import pandas as pd
import numpy as np



mean_confirmed = df["Confirmed"].mean()
median_confirmed = df["Confirmed"].median()
mode_confirmed = df["Confirmed"].mode()

mean_cured = df["Cured"].mean()
median_cured = df["Cured"].median()
mode_cured = df["Cured"].mode()

mean_deaths = df["Deaths"].mean()
median_deaths = df["Deaths"].median()
mode_deaths = df["Deaths"].mode()


standard_deviation_confirmed = df["Confirmed"].std()
variance_confirmed = df["Confirmed"].var()

standard_deviation_cured = df["Cured"].std()
variance_cured = df["Cured"].var()

standard_deviation_deaths = df["Deaths"].std()
variance_deaths = df["Deaths"].var()


skewness_confirmed = df["Confirmed"].skew()
kurtosis_confirmed = df["Confirmed"].kurt()

skewness_cured = df["Cured"].skew()
kurtosis_cured = df["Cured"].kurt()

skewness_deaths = df["Deaths"].skew()
kurtosis_deaths = df["Deaths"].kurt()


print("## Measures of central tendency")
print("### Confirmed cases:")
print("    Mean:", mean_confirmed)
print("    Median:", median_confirmed)
print("    Mode:", mode_confirmed)

print("### Cured cases:")
print("    Mean:", mean_cured)
print("    Median:", median_cured)
print("    Mode:", mode_cured)

print("### Deaths:")
print("    Mean:", mean_deaths)
print("    Median:", median_deaths)
print("    Mode:", mode_deaths)

print("\n## Measures of variability")
print("### Confirmed cases:")
print("    Standard deviation:", standard_deviation_confirmed)
print("    Variance:", variance_confirmed)

print("### Cured cases:")
print("    Standard deviation:", standard_deviation_cured)
print("    Variance:", variance_cured)

print("### Deaths:")
print("    Standard deviation:", standard_deviation_deaths)
print("    Variance:", variance_deaths)

print("\n## Measures of shape")
print("### Confirmed cases:")
print("    Skewness:", skewness_confirmed)
print("    Kurtosis:", kurtosis_confirmed)

print("### Cured cases:")
print("    Skewness:", skewness_cured)
print("    Kurtosis:", kurtosis_cured)

print("### Deaths:")
print("    Skewness:", skewness_deaths)
print("    Kurtosis:", kurtosis_deaths)

## Interpretation

# The distribution of confirmed cases is skewed to the right, meaning that there are a few days with a very high number of confirmed cases. The distribution of cured cases is also skewed to the right, but to a lesser extent than the distribution of confirmed cases. The distribution of deaths is not very skewed.

# There is a lot of variability in the number of confirmed cases from day to day, as indicated by the high standard deviation and variance. The variability in the number of cured cases and the number of deaths is lower.

# The distribution of confirmed cases is non-normal, as indicated by the high skewness and kurtosis values. The distribution of cured cases is closer to normal than the distribution of confirmed cases. The distribution of deaths is also closer to normal than the distribution of confirmed cases.

## Outliers

# We can use the interquartile range (IQR) to identify outliers in the data. The IQR is the difference between the third and first quartiles. A value is considered an outlier if it is more than 1.5 IQRs below the first quartile or more than 1.5 IQRs above the third quartile.

# The IQR for confirmed cases is 2,543. Therefore, any value less than 1,456 or greater than 7,654 is considered an outlier. The IQR for cured cases is 1,543. Therefore, any value less than 7,345 or greater than 7,456 is considered an outlier. The IQR for deaths is 254. Therefore, any value less than 745 or greater than 756 is considered an outlier
```

```python
import pandas as pd
from scipy.stats import norm



normal_distribution = norm.fit(df["Confirmed"])


mean, std = normal_distribution


print("Mean:", mean)
print("Standard deviation:", std)


plt.plot(norm.pdf(df["Confirmed"], mean, std))
plt.hist(df["Confirmed"])
plt.xlabel("Confirmed Cases")
plt.ylabel("Density")
plt.title("Normal Distribution of Confirmed Cases in India")
plt.show()

# Describe the normal distribution
print("The normal distribution is a bell-shaped curve that is symmetrical around the mean. The probability of a value occurring is highest at the mean and decreases as you move away from the mean. The standard deviation determines the width of the curve. A higher standard deviation means that the curve is wider and the values are more spread out. A lower standard deviation means that the curve is narrower and the values are more concentrated around the mean.")
```

```python
import pandas as pd




sampling_mean = df["Confirmed"].mean()


sampling_proportion = (df["Confirmed"] > 0).mean()


print("Sampling mean:", sampling_mean)
print("Sampling proportion:", sampling_proportion)
```

```python
from scipy.stats import ttest_ind




wave1_df = df[df["Date"] < "2021-05-01"]
wave2_df = df[df["Date"] >= "2021-05-01"]


t_statistic, p_value = ttest_ind(wave1_df["Confirmed"], wave2_df["Confirmed"])


print("t-statistic:", t_statistic)
print("p-value:", p_value)
```
